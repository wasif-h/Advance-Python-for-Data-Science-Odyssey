# Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = [1,2,3,4,5]
y = [5,12,10,34,22]

plt.plot(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Basic Line Plot")

plt.show()
```

```python
# customized line plot 

x = [1,2,3,4,5]
y = [5,12,10,34,22]

plt.plot(x,y,color="red", linestyle="dotted",linewidth="2",marker='o',markersize='10')
plt.grid(True)
```

```python
# Subplotting 

x = np.arange(1,6,1)
y1 = np.random.randint(1,26,size=[5])
y2 = np.random.randint(1,26,size=[5])
y3 = np.random.randint(1,26,size=[5])
y4 = np.random.randint(1,26,size=[5])
```

```python
plt.figure(figsize=(15,10))

plt.subplot(2,2,1)
plt.plot(x,y1,color="red",marker="o")
plt.title("Plot 1")

plt.subplot(2,2,2)
plt.plot(x,y2,color="green",marker="o")
plt.title("Plot 2")

plt.subplot(2,2,3)
plt.plot(x,y3,color="blue",marker="o")
plt.title("Plot 3")

plt.subplot(2,2,4)
plt.plot(x,y4,color="black",marker="o")
plt.title("Plot 4")
```

```python
# Bar Plot

x = ['a','b','c','d','e']
y = [56,23,67,21,33]

plt.bar(x,y,color="purple")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Stat")
```

```python
data = np.random.randint(0,100,size=[30])
data
```

```python
plt.hist(data,bins=7,color="orange",edgecolor="black")
```

```python
x = np.random.randint(-6,6,size=[30])
y = np.random.randint(-6,6,size=[30])
plt.scatter(x,y,marker="x")
plt.grid(True)
```

```python
data = [20,39,11,56,23]
labels = ['a','b','c','d','e']
explode = [0.2,0,0,0,0]

plt.pie(data,labels=labels, explode=explode,shadow=True,autopct="%1.1f%%");
```
