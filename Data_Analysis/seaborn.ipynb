# Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
print(tips.info())
print(tips.describe())
print(tips.sample(5))
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
```

```python
print(tips.sample(10))
```

```python
# day wise total sale 

total_sale_of_Day = tips.groupby('day')['total_bill'].sum().reset_index()
total_sale_of_Day
```

```python
sns.barplot(data=total_sale_of_Day,x = 'day',y='total_bill')
```

```python
tips.head()
```

```python
sns.lineplot(data = tips, x = "size",y = "total_bill")
```

```python
# automatically categorical bar plot, it can find avg based on grp 

sns.barplot(x = "day", y= "total_bill",data = tips)
```

```python
sns.boxplot(x = 'day',y = 'total_bill',data=tips)
```

```python
sns.violinplot(x = 'day',y = 'total_bill',data=tips)
```

```python
sns.histplot(tips['total_bill'],kde=True)
```

```python
sns.kdeplot(tips['total_bill'],fill=True)
```

```python
sns.pairplot(tips)
```

```python
df = tips[["total_bill",'tip','size']].corr()
df.head()
```

```python
sns.heatmap(df,annot=True,cmap="coolwarm")
```
